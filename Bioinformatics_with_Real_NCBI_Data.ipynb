**4-Download a real sequence: mitochondrial sequence of *Apis mellifera***

```python
handle = Entrez.efetch(
    db="nucleotide",
    id="NC_001566",
    rettype="fasta",
    retmode="text"
)

record = SeqIO.read(handle, "fasta")

print(record.id)
print(len(record.seq))
```

**5-Calculate characteristics**

```python
seq = str(record.seq)

length = len(seq)

gc = (
    seq.count("G")
    + seq.count("C")
) / length * 100

print("Length:", length)
print("GC %:", round(gc,2))
```

**7-Build DataFrame**

```python
import pandas as pd

accessions = [
    "NC_001566",
    "NC_015770",
    "NC_028009"
]

data = []

for acc in accessions:

    handle = Entrez.efetch(
        db="nucleotide",
        id=acc,
        rettype="fasta",
        retmode="text"
    )

    record = SeqIO.read(
        handle,
        "fasta"
    )

    seq = str(record.seq)

    gc = (
        seq.count("G")
        + seq.count("C")
    ) / len(seq) * 100

    data.append([
        acc,
        len(seq),
        gc
    ])

df = pd.DataFrame(
    data,
    columns=[
        "Accession",
        "Length",
        "GC_Content"
    ]
)

df
```
